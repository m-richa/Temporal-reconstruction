```python
rotate_1 = trimesh.transformations.rotation_matrix(angle=np.radians(90), direction=[0, 0, -1])
rotate_2 = trimesh.transformations.rotation_matrix(angle=np.radians(180), direction=[0,1,0])
extrinsic = torch.tensor(rotate_1[:3, :3] @ rotate_2[:3, :3]).float()
```

```python
mask = torch.where(torch.tensor(depth_curr[0])>0)
y, x = mask
K = torch.tensor(K)
im_points = torch.stack((x, y, torch.ones(len(x))), 0)
print(im_points.shape)
world_points = (torch.linalg.inv(extrinsic) @ torch.linalg.inv(K) @ (im_points * depth_curr[0][mask])).cpu().numpy()
world_points_next = (world_points.T + scf[0][mask]).T
world_points_next.shape
```

```python
rgb_path = os.makedirs(root_gt+'/rgb', exist_ok=True )
depth_path = os.makedirs(root_gt+'/depth', exist_ok=True )
scf_path = os.makedirs(root_gt+'/scf', exist_ok=True )
```

```python
for seq in seqs:
    print('Sequence: '+str(seq))
    os.makedirs(root_gt+'/rgb/'+str(seq), exist_ok=True)
    os.makedirs(root_gt+'/depth/'+str(seq), exist_ok=True)
    os.makedirs(root_gt+'/scf/'+str(seq), exist_ok=True)
    
    path_seq = root + '/' + str(seq)
    
    for i in range(4):

        rgb_out_dir = root_gt+'/rgb/'+str(seq)+'/Frame{num:03d}'.format(num=i)
        depth_out_dir = root_gt+'/depth/'+str(seq)+'/Frame{num:03d}'.format(num=i)
        scf_out_dir = root_gt+'/scf/'+str(seq)+'/Frame{num:03d}'.format(num=i)
        
        os.makedirs(rgb_out_dir, exist_ok=True)
        os.makedirs(depth_out_dir, exist_ok=True)
        os.makedirs(scf_out_dir, exist_ok=True)
        
        
        path = path_seq + '/Frame{num:03d}'.format(num=i) + '/human/body.obj'
        
        intersections, ray_param, ray_intersect_param, scene = get_peeled_intersections(root, path)
        
        get_rgb_peels(root, path, intersections, ray_param, ray_intersect_param, scene, rgb_out_dir)
        get_depth_peels(root, path, intersections, ray_param, ray_intersect_param, scene, depth_out_dir)
        
        while(i<299):
            path_next = path_seq + '/Frame{num:03d}'.format(num=i+1) + '/human/body.obj'
            _, _ = get_scf_peels(root, path, path_next,
                             intersections, ray_param,
                             ray_intersect_param, scene, scf_out_dir)
        
        
    
```
